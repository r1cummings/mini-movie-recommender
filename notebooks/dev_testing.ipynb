```python
import pandas as pd
import os
```

```python
r_cols = ['user_id', 'movie_id', 'rating']
ratings = pd.read_csv('data/u.data', sep='\t', names=r_cols, usecols=range(3), encoding="ISO-8859-1")

m_cols = ['movie_id', 'title']
movies = pd.read_csv('data/u.item', sep='|', names=m_cols, usecols=range(2), encoding="ISO-8859-1")

ratings = pd.merge(movies, ratings)
```

```python
test_movies = movies.copy()
test_movies = test_movies.drop(['movie_id'],axis=1)
test_movies['title'] = [i.replace('(','').replace(')','') for i in test_movies['title'].values]
test_movies.to_csv('movie_names.csv')
```

```python
ratings['title'] = [i.replace('(','').replace(')','') for i in ratings['title'].values]
```

```python
ratings.tail(10)
```

```python
user_ratings = ratings.pivot_table(index=['user_id'],columns=['title'],values='rating')
user_ratings.head()
```

```python
corrMatrix = user_ratings.corr(method='pearson', min_periods=100)
corrMatrix.head()
```

```python
myRatings = user_ratings.loc[0].dropna()
myRatings
```

```python
myRatings.index
```

```python
# change data to be a list of
import numpy as np

x = np.array([5])
your_fav_movies = input("What is/are your all time favorite movie(s)?:\n")
your_fav_movies = your_fav_movies.split(';')
movie_lst = [i.lstrip() for i in your_fav_movies]
myRatings = pd.Series(data = np.repeat(x, [len(movie_lst)], axis=0), index = movie_lst)
myRatings
```

```python
len(myRatings.index)
```

```python
# have the user input their fave movies seperated by comma and in strings with year and then that becomes myRatings
# myRatings = ['Empire Strikes Back, The (1980)']
simCandidates = pd.Series()
for i in range(0, len(myRatings.index)):
    # Retrieve similar movies to this one that I rated
    sims = corrMatrix[myRatings.index[i]].dropna()
    # Now scale its similarity by how well I rated this movie
    sims = sims.map(lambda x: x * myRatings[i])
    # Add the score to the list of similarity candidates
    simCandidates = simCandidates.append(sims)
    
#Glance at our results so far:
simCandidates.sort_values(inplace = True, ascending = False)
```

```python
simCandidates = simCandidates.groupby(simCandidates.index).sum()
simCandidates.sort_values(inplace = True, ascending = False)
filteredSims = simCandidates.drop(myRatings.index)
```

```python
print('Your top 3 recommended movies are:\n')

print('\n'.join(list(filteredSims.index[0:3].values)))
```
